# Useful Functions in Lk
* flatten(): removes long term trends (via Savitzkyâ€“Golay filter)
* remove_outliers(): removes outliers (simple sigma clipping)
* remove_nans(): removes infinite/NaN values
* fold(): fold data at particular period

  ex: folded_lc = flat_lc.fold(period = (time in days))

  one line version:
                          lcf[0].flatten(window_length=401).fold(period=3.690621).bin(time_bin_size=0.01).plot();
* bin(): reduce time resolution of array, takes average of each bin
* .interact_bls(): identify periodic transit signals in light curves by manually selecting period & duration of signal
